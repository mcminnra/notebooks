```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as stats
```

## Beta Distribution

The Beta Distribution is a probability distribution of probabilities. In other words, it models ratios.

```python
# What's the rating distribution from the ratio of likes and dislikes?
# Answer - Largely between 81%  - 85%

a = 147  # Likes
b = 30  # Dislikes

x = np.linspace(0, 1, 1000)  # Some values from 0 to 1
y = stats.beta.pdf(x, a, b)  # Sample from distribution

# Summary Statistics (aka The First 4 Moments + Std. Dev.)
mean, var, skew, kurt = stats.beta.stats(a, b, moments='mvsk')
std_dev = np.sqrt(var)

print(f'Mean: {mean}')
print(f'Variance: {var}')
print(f'Std. Dev.: {std_dev}')
print(f'Skew: {skew}')
print(f'Kurtosis.: {kurt}')

# Plot Distribution
plt.figure(figsize=(14, 8))
plt.plot(x, y, 'r-', lw=5, alpha=0.6, label='beta pdf')
plt.title(f'Beta')
plt.grid()
plt.show()
```

## Exponential Distribution

Exponential Distribution models wait time until first event.

```python
# How long to wait until a NBA Basketball Game for a single team?
# Answer - About 4 days

lam = 82  # Rate of Events following Poisson dist. - 82 games in a year (Regular Season)
theta = 1/lam  # Scale

x = np.linspace(0, 1, 1000)  # Some values
y = stats.expon.pdf(x, scale=theta)  # Sample from distribution

# Summary Statistics (aka The First 4 Moments + Std. Dev.)
mean, var, skew, kurt = stats.expon.stats(scale=theta, moments='mvsk')
std_dev = np.sqrt(var)

print(f'Mean: {mean}')
print(f'Variance: {var}')
print(f'Std. Dev.: {std_dev}')
print(f'Skew: {skew}')
print(f'Kurtosis.: {kurt}')

# Plot Distribution
plt.figure(figsize=(14, 8))
plt.plot(x, y, 'r-', lw=5, alpha=0.6)
plt.title(f'Exponential')
plt.grid()
plt.show()
```

## Gamma Distribution

The Gamma Distribution models the wait time until the k-th event.

```python
# Time until 1 million covid deaths
# Answer - Roughly 22.8 years if current rates hold.

k = 1000000  # Num. of Events - 1,000,000 covid deaths
lam = 842  # Rate of Events following Poisson dist. - 842 deaths a week
theta = 1/lam  # Scale

x = np.linspace(1100, 1200, 1000)  # Some values
y = stats.gamma.pdf(x, k, scale=theta)  # Sample from distribution

# Summary Statistics (aka The First 4 Moments + Std. Dev.)
mean, var, skew, kurt = stats.gamma.stats(k, scale=theta, moments='mvsk')
std_dev = np.sqrt(var)

print(f'Mean: {mean}')
print(f'Variance: {var}')
print(f'Std. Dev.: {std_dev}')
print(f'Skew: {skew}')
print(f'Kurtosis.: {kurt}')

# Plot Distribution
plt.figure(figsize=(14, 8))
plt.plot(x, y, 'r-', lw=5, alpha=0.6)
plt.title(f'Gamma')
plt.grid()
plt.show()
```

## Logistic Distribution

The Logistic Distribution resembles the Normal Distribution with heavier tails (higher kurtosis), meaning events at the fringes of the distribution have a higher chance of happening than a Normal Distribution. Fun fact, the CDF of the Logistic Distribution is the Logistic Function, or otherwise known as Sigmoid Activation.  

```python
mu = 70  # Mean (Location / Loc)
s = 3 # scale (This is not Std. Dev.) 

x = np.linspace(40, 100, 1000)  # Some values
y = stats.logistic.pdf(x, mu, s)  # Sample from distribution

# Summary Statistics (aka The First 4 Moments + Std. Dev.)
mean, var, skew, kurt = stats.logistic.stats(mu, s, moments='mvsk')
std_dev = np.sqrt(var)

print(f'Mean: {mean}')
print(f'Variance: {var}')
print(f'Std. Dev.: {std_dev}')
print(f'Skew: {skew}')
print(f'Kurtosis.: {kurt}')

# Plot Distribution
plt.figure(figsize=(14, 8))
plt.plot(x, y, 'r-', lw=5, alpha=0.6)
plt.title(f'Logistic')
plt.grid()
plt.show()
```

## Normal Distribution (or Gaussian Distribution)

Intuitively, the normal distribution is "nice" enough that we expect it to occur naturally. This intuition is formalized by the Central Limit Theorem. Many patterns found naturally follow a normal distribution.

```python
# Modeling height on american human males

mu = 70  # Height (in inches)
sigma = 3 # Std. Dev. Height (in inches)

x = np.linspace(40, 100, 1000)  # Some values
y = stats.norm.pdf(x, mu, sigma)  # Sample from distribution

# Summary Statistics (aka The First 4 Moments + Std. Dev.)
mean, var, skew, kurt = stats.norm.stats(mu, sigma, moments='mvsk')
std_dev = np.sqrt(var)

print(f'Mean: {mean}')
print(f'Variance: {var}')
print(f'Std. Dev.: {std_dev}')
print(f'Skew: {skew}')
print(f'Kurtosis.: {kurt}')

# Plot Distribution
plt.figure(figsize=(14, 8))
plt.plot(x, y, 'r-', lw=5, alpha=0.6)
plt.title(f'Normal')
plt.grid()
plt.show()
```

## Half-Normal Distribution

Half-Normal Distribution, and other "Half" distributions only give half of the specified distribution. They are used to model situations where the left side of the distribution doesn't make sense. Most often we see these where we want to model a Normal(mu,sigma) where mu=0, and negative values can't exist, therefore the probability of a negative value is 0. 

```python
# Modeling height on american human males

mu = 70  # Height (in inches)
sigma = 3 # Std. Dev. Height (in inches)

x = np.linspace(40, 100, 1000)  # Some values
y = stats.halfnorm.pdf(x, mu, sigma)  # Sample from distribution

# Summary Statistics (aka The First 4 Moments + Std. Dev.)
mean, var, skew, kurt = stats.halfnorm.stats(mu, sigma, moments='mvsk')
std_dev = np.sqrt(var)

print(f'Mean: {mean}')
print(f'Variance: {var}')
print(f'Std. Dev.: {std_dev}')
print(f'Skew: {skew}')
print(f'Kurtosis.: {kurt}')

# Plot Distribution
plt.figure(figsize=(14, 8))
plt.plot(x, y, 'r-', lw=5, alpha=0.6)
plt.title(f'Half-Normal')
plt.grid()
plt.show()
```

## Truncated Normal Distribution

Similar to a Half distribution, Truncated Normal, and other Truncated Distributions, are distributions where the values can only happen within a certain range or bound. We use these when we have an intuition on the range a Random Variable can take. 

```python
# Modeling height on american human males

mu = 70  # Height (in inches)
sigma = 3 # Std. Dev. Height (in inches)
a = (65 - mu)/sigma  # Lower Bound (in # of Std. Devs.)
b = (75 - mu)/sigma  # Upper Bound (in # of Std. Devs.)

x = np.linspace(40, 100, 1000)  # Some values
y = stats.truncnorm.pdf(x, a, b, mu, sigma)  # Sample from distribution

# Summary Statistics (aka The First 4 Moments + Std. Dev.)
mean, var, skew, kurt = stats.truncnorm.stats(a, b, mu, sigma, moments='mvsk')
std_dev = np.sqrt(var)

print(f'Mean: {mean}')
print(f'Variance: {var}')
print(f'Std. Dev.: {std_dev}')
print(f'Skew: {skew}')
print(f'Kurtosis.: {kurt}')

# Plot Distribution
plt.figure(figsize=(14, 8))
plt.plot(x, y, 'r-', lw=5, alpha=0.6)
plt.title(f'Truncated Normal')
plt.grid()
plt.show()
```

## Pareto Distribution

Pareto Distribution models the situation where a large potion of outcomes are due to a small fraction of the causes. The Pareto Principle, or the "80-20" rule, is an example of a Pareto(1.16) where 80% of the outcomes are due to 20% of the causes. The principle is famously seen in wealth distribution where a large portion of wealth is held by a small fraction of the population.

MLE Estimation for this distribution is suggested if you suspect your data is Pareto distributed. The loc and scale params are a little unintuitive. 

```python
q = .2
p = .8
a = np.log(q) / np.log(q/p)  # a = 1.16 # shape or "Pareto Index" when used for wealth 

x = np.linspace(0, 10, 1000)  # Some values
y = stats.pareto.pdf(x, a)  # Sample from distribution

# Summary Statistics (aka The First 4 Moments + Std. Dev.)
mean, var, skew, kurt = stats.pareto.stats(a, moments='mvsk')
std_dev = np.sqrt(var)

print(f'Mean: {mean}')
print(f'Variance: {var}')
print(f'Std. Dev.: {std_dev}')
print(f'Skew: {skew}')
print(f'Kurtosis.: {kurt}')

# Plot Distribution
plt.figure(figsize=(14, 8))
plt.plot(x, y, 'r-', lw=5, alpha=0.6)
plt.title(f'Pareto')
plt.grid()
plt.show()
```

## Poisson Distribution

"[The Poisson Distribution] expresses the probability of a given number of events occurring in a fixed interval of time or space if these events occur with a known constant mean rate and independently of the time since the last event." - Wikipedia

```python
# Modeling the rate of meals in a day
# Answer - Largely between 2 and 4 (duh)

lam = 3  # Rate of Events following Poisson dist. - 3 Meals in a day

x = list(range(10))  # Some values
y = stats.poisson.pmf(x, lam)  # Sample from distribution

# Summary Statistics (aka The First 4 Moments + Std. Dev.)
mean, var, skew, kurt = stats.poisson.stats(lam, moments='mvsk')
std_dev = np.sqrt(var)

print(f'Mean: {mean}')
print(f'Variance: {var}')
print(f'Std. Dev.: {std_dev}')
print(f'Skew: {skew}')
print(f'Kurtosis.: {kurt}')

# Plot Distribution
plt.figure(figsize=(14, 8))
plt.plot(x, y, 'r-', lw=5, alpha=0.6)
plt.title(f'Poisson')
plt.grid()
plt.show()
```

## Student's T Distribution

Student's T is used to estimate the mean of a normally distributed population when the sample size is small and the population Std. Dev. is unknown. As an example, you might collect 10 samples of something. You know the sample mean and sample variance, but you don't know the population mean or the population variance. Student's T is, given the number of samples, what is the distribution of how many sample Std. Deviations the population mean is from the sample mean. This T distribution allows you to construct a Confidence Interval for your sample mean. The more samples (or degrees of freedom), the more "constrained" the T distribution becomes.

The T Distribution is a distribution of T Statistics. 

```python
n = 10  # Samples
v = n-1  # Degrees on freedom (n-k)

x = np.linspace(-10, 10, 1000)  # Some values
y = stats.t.pdf(x, v)  # Sample from distribution

# Summary Statistics (aka The First 4 Moments + Std. Dev.)
mean, var, skew, kurt = stats.t.stats(v, moments='mvsk')
std_dev = np.sqrt(var)

print(f'Mean: {mean}')
print(f'Variance: {var}')
print(f'Std. Dev.: {std_dev}')
print(f'Skew: {skew}')
print(f'Kurtosis.: {kurt}')

# Plot Distribution
plt.figure(figsize=(14, 8))
plt.plot(x, y, 'r-', lw=5, alpha=0.6)
plt.title(f'Student\'s T')
plt.grid()
plt.show()
```
