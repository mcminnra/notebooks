# Beta Distribution

Beta Distribution is used for modeling percentages or proportions.

Beta Distribution takes two params, alpha and beta. Alpha can be interpreted as true/successes/hits and Beta as false/failures/misses

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import beta
```

```python
# Let's take a Season-Long Batting Average as an example - 27% or .27

# Roughly equals .27 and generates a distribution between .2 and .35
a = 81  # Hits
b = 219  # Misses

mean, var, skew, kurt = beta.stats(a, b, moments='mvsk')

print(f'Mean:\t\t {mean}')
print(f'Variance:\t {var}')
print(f'Skew:\t\t {skew}')
print(f'Kurtosis:\t {kurt}')
```

```python
print(f'Value at 5th percentile:\t {beta.ppf(.05, a, b)}')
print(f'Value at 95th percentile:\t {beta.ppf(.95, a, b)}')
```

```python
plt.figure(figsize=(14, 8))
x = np.linspace(beta.ppf(0, a, b), beta.ppf(1, a, b), 100)
plt.plot(x, beta.pdf(x, a, b), 'r-', lw=5, alpha=0.6, label='beta pdf')
plt.grid()
plt.xlim(0, 1)
plt.show()
```

```python
# Updating
#beta(a_0 + hits, b_0 + misses)

# Let's say they added 100 hits and 200 misses - Buddy just started killing it on the mound
a_1 = a + 100
b_1 = b + 200

plt.figure(figsize=(14, 8))
x = np.linspace(beta.ppf(0, a, b), beta.ppf(1, a, b), 100)
x_1 = np.linspace(beta.ppf(0, a_1, b_1), beta.ppf(1, a_1, b_1), 100)

plt.plot(x, beta.pdf(x, a, b), 'r-', lw=5, alpha=0.6, label='beta pdf')
plt.plot(x_1, beta.pdf(x_1, a_1, b_1), 'b-', lw=5, alpha=0.6, label='beta pdf Updated')
plt.grid()
plt.xlim(0, 1)

print(f'Prior Mean:\t {beta.stats(a, b, moments="m")}')
print(f'Updated Mean:\t {beta.stats(a_1, b_1, moments="m")}')
```

The larger the alpha and beta, the "stronger" the evidence is - meaning the distribution will have a lower variance.
